# Alucar - Analisando as vendas

```python
import pandas as pd
```

```python
alucar = pd.read_csv('alucar.csv')
```

```python
alucar.shape
```

```python
alucar.isna().value_counts()
```

```python
alucar['mes'] = pd.to_datetime(alucar['mes'])
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
sns.lineplot(x='mes',y='vendas',data=alucar)
```

```python
sns.set_palette('Accent')
sns.set_style('darkgrid')
ax = sns.lineplot(x='mes',y='vendas',data=alucar)
ax.figure.set_size_inches(12,6)
ax.set_title('Vendas da Alucar de 2017-2018',loc='left',fontsize=20)
ax.set_xlabel('Ms',fontsize=14)
ax.set_ylabel('Vendas (un)',fontsize=14)
```

```python
alucar['aumento']=alucar['vendas'].diff()
```

```python
alucar.head()
```

```python
sns.set_palette('Accent')
sns.set_style('darkgrid')
ax = sns.lineplot(x='mes',y='aumento',data=alucar)
ax.figure.set_size_inches(12,6)
ax.set_title('Aumento das vendas da Alucar de 2017-2018',loc='left',fontsize=20)
ax.set_xlabel('Ms',fontsize=14)
ax.set_ylabel('Aumento (R$)',fontsize=14)
```

```python
def plotar(dados,x,y,titulo,label_x,label_y):
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    ax = sns.lineplot(x=x,y=y,data=dados)
    ax.figure.set_size_inches(12,6)
    ax.set_title(titulo,loc='left',fontsize=20)
    ax.set_xlabel(label_x,fontsize=14)
    ax.set_ylabel(label_y,fontsize=14)
    ax =ax
```

```python
plotar(alucar,alucar.mes,alucar.aumento,'Aumento das vendas da Alucar 2017-2018','Ms','Aumento (R$)')
```

```python
alucar['aceleracao'] = alucar['aumento'].diff()
```

```python
plotar(alucar,alucar.mes,alucar.aceleracao,'Acelerao das vendas da Alucar 2017-2018','Ms','Acelerao (R$)')
```

```python
plt.figure(figsize = (16,12))
ax = plt.subplot(3,1,1)
ax.set_title('Anlise da Alucar de 2017-2018',fontsize=18,loc='left')
sns.lineplot(x='mes',y='vendas',data=alucar)
plt.subplot(3,1,2)
sns.lineplot(x='mes',y='aumento',data=alucar)
plt.subplot(3,1,3)
sns.lineplot(x='mes',y='aceleracao',data=alucar)
ax = ax
```

```python
def plot_comparacao(x, y1, y2, y3, dados, titulo):
    plt.figure(figsize = (16,12))
    ax = plt.subplot(3,1,1)
    ax.set_title(titulo,fontsize=18,loc='left')
    sns.lineplot(x=x,y=y1,data=dados)
    plt.subplot(3,1,2)
    sns.lineplot(x=x,y=y2,data=dados)
    plt.subplot(3,1,3)
    sns.lineplot(x=x,y=y3,data=dados)
    ax = ax
```

```python
plot_comparacao('mes','vendas','aumento','aceleracao',alucar, 'Anlise das vendas da Alucar 2017-2018')
```

```python
from pandas.plotting import autocorrelation_plot
```

```python
ax = plt.figure(figsize=(16,12))
ax.suptitle('Correlao das vendas', fontsize=20, x=0.25, y=0.95)
autocorrelation_plot(alucar['vendas'])
ax = ax
```

```python
ax = plt.figure(figsize=(16,12))
ax.suptitle('Correlao do aumento', fontsize=20, x=0.25, y=0.95)
autocorrelation_plot(alucar['aumento'][1:])
ax = ax
```
