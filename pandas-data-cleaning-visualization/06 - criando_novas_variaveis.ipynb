## Criando novas variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio']
```

```python
dados.head()
```

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head()
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
```

```python
dados.head()
```

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto']/dados['Area']).round(2)
```

```python
dados.head()
```

```python
casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados
```

## Excluindo Variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado','Valor m2','Valor Bruto','Valor Bruto m2']])
```

```python
dados_aux.head()
```

```python
dados.drop(['Unnamed: 0','Valor Bruto','Valor Bruto m2'], axis = 1, inplace = True)
```
