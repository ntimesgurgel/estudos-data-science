# Tradução

```python
gorjetas.columns
```

```python
renomear = {
    'total_bill':'valor_da_conta',
    'tip':'gorjeta',
    'dessert':'sobremesa',
    'day':'dia',
    'time':'tempo',
    'size':'tamanho'
}
```

```python
gorjetas = gorjetas.rename(columns = renomear)
```

```python
gorjetas
```

```python
gorjetas.sobremesa.unique()
```

```python
sim_nao = {
    'No':'N',
    'Yes':'S'
}
dias = {
    'Sun':'Dom',
    'Sat':'Sab', 
    'Thur':'Qui', 
    'Fri':'Sex'
}
tempos = {
    'Dinner':'Jantar',
    'Lunch':'Almoço'
}
```

```python
gorjetas.sobremesa = gorjetas.sobremesa.map(sim_nao)
gorjetas.dia = gorjetas.dia.map(dias)
```
